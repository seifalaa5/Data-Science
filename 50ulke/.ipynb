# Checking Countries With The Highest Income

```python
sns.scatterplot(x='National Income',y="Countries",data=df)
```

```python
df.sort_values(by='National Income',ascending=False)[["Countries",'National Income']].head()
```

##### Checking Countries With The Lowest Income

```python
df.sort_values(by='National Income',ascending=False)[["Countries",'National Income']].tail()
```

```python
sns.scatterplot(x='Growth',y="Countries",data=df)
```

##### Highest Growth
